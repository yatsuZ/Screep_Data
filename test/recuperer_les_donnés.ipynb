```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.allocine.fr/films/'#l' url du lien que je veux scraper

response = requests.get(url)#je recupere la page et elle est en requeste

#soup = au body de ma page interprete en html
soup = BeautifulSoup(response.text, 'html.parser')
all_presentations = soup.find_all("div", class_="card entity-card entity-card-list cf")
presentation = soup.find("div", class_="card entity-card entity-card-list cf")
i = 0

affinage = soup.find("nav", "pagination cf")
print(affinage.find("a", "xXx.button.button-md.button-primary-full.button-right"))
print("\n\n\n")
#print(presentation.find("div", "meta-body-item meta-body-actor").text[5:])
#print(all_presentations)
#ptiInfo = presentation.find("div", "meta-body-item meta-body-info").text
#date, time, categories = extract_time_and_categories(ptiInfo)
#print(date)
#print(time)
#categories = " ".join(categories)
#print(categories)
#print(soup.find("a", "meta-title-link"))
#print(presentation.find("a", "meta-title-link").text)
#print(all_presentations)
########################################################################################
def extract_time_and_categories(text):
    parts = text.split("\n")
    date = parts[1].strip()
    time = parts[3].strip()
    categories = parts[5:]
    return date, time, categories
##########################################################################################
def extract_Acteurs(text):
    text = text[5:]
    realisateur = ""
    for i in text :
        if (i != "\n"):
            realisateur += i
        else :
            realisateur += " "
    return realisateur[: -1]
################################################################################################


for presentations in all_presentations:
    notes = presentation.find_all("span", "stareval-note")
    notePresse = notes[0].text
    notePublique = notes[1].text
#############################################
    synopsys = presentations.find("div", "content-txt").text
############################################
    acteurs = presentations.find("div", "meta-body-item meta-body-actor").text
    acteurs = extract_Acteurs(acteurs)
#############################################""
    realisateurBalise = presentations.find("div", "meta-body-item meta-body-direction")
    realisateurSpan = realisateurBalise.find_all("span")[1:]
    realisateurA = realisateurBalise.find_all("a")
    realisateur = ""
    for nom in realisateurA :
        realisateur += nom.text + ", "
    for nom in realisateurSpan :
        realisateur += nom.text + ", "
    realisateur = realisateur[: -2]
###############################
    date, time, categories = extract_time_and_categories(presentation.find("div", "meta-body-item meta-body-info").text)
    categories = " ".join(categories)
################################
    photo = presentations.find("img", "thumbnail-img")
    key = photo.get("data-src")
    if key is not None:
        photo = presentations.find("img", "thumbnail-img")['data-src']
    else:
        photo = presentations.find("img", "thumbnail-img")['src']
################################
    titre = presentations.find("a", "meta-title-link").text
    print("ID = " + str(i),"\ntitre = " + titre, "\nimage = " + photo, "\ndate = " + date, "\ntime = " + time, "\ncategorie = " + categories)
    print("Realisateur = " + realisateur, "\nActeur = " + acteurs, "\nSynopsys = " + synopsys[:-1], "\nNotePresse = " + notePresse)
    print("NotePublique = " + notePublique, "\n-------------------")
################################
    i = i + 1
```
